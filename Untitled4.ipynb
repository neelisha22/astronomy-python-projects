```python
# Core libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# NetCDF reader
import xarray as xr

# Signal processing
from scipy.signal import savgol_filter
from scipy.fft import rfft, rfftfreq
# Wavelet analysis
import pywt

from PyEMD import EMD
import numpy as np
import matplotlib.pyplot as plt
```

```python
# open GOES dataset
ds = xr.open_dataset("/media/hp/TOSHIBA EXT/sci_xrsf-l2-flx1s_g18_d20260201_v2-2-1.nc")

# extract variables
time = pd.to_datetime(ds['time'].values)
xrsa = ds['xrsa_flux'].values   # 0.5–4 Å
xrsb = ds['xrsb_flux'].values   # 1–8 Å

plt.figure(figsize=(12,5))

plt.plot(time, xrsa, label="0.5–4 Å")
plt.plot(time, xrsb, label="1–8 Å")

plt.yscale("log")

plt.xlabel("Time")
plt.ylabel("Flux (W/m²)")
plt.title("GOES Full-Day X-ray Light Curve (Raw Data, Log Scale)")

plt.legend()
plt.grid(alpha=0.3)

plt.show()
```

```python
from scipy.signal import find_peaks

# thresholds
c_min = 1e-6
c_max = 1e-5

# detect peaks with stronger conditions
peaks, properties = find_peaks(
    xrsb,
    height=1e-6,       # C flare threshold
    prominence=5e-7,   # smaller prominence
    distance=120       # avoid multiple peaks in same flare
)
# select C-class only
c_peaks = [p for p in peaks if xrsb[p] < c_max]

print("Number of C-class flares:", len(c_peaks))
```

```python
plt.figure(figsize=(12,5))

plt.plot(time, xrsb, label="1–8 Å")

plt.scatter(time[c_peaks], xrsb[c_peaks], color='red', label='C flares')

plt.yscale("log")
plt.xlabel("Time")
plt.ylabel("Flux (W/m²)")
plt.title("Detected C-class flares")

plt.legend()
plt.show()
```

```python
flare_windows = []

for p in c_peaks:

    peak_time = time[p]

    start = peak_time - np.timedelta64(10, 'm')
    end   = peak_time + np.timedelta64(10, 'm')

    mask = (time >= start) & (time <= end)

    t_flare = time[mask]
    flux_flare = xrsb[mask]

    flare_windows.append((t_flare, flux_flare))
```

```python
import matplotlib.pyplot as plt

n = len(flare_windows)

fig, axes = plt.subplots(2, 6, figsize=(22,8))

axes = axes.flatten()

for i, (t_flare, flux_flare) in enumerate(flare_windows):

    axes[i].plot(t_flare, flux_flare)
    axes[i].set_yscale("log")

    axes[i].set_title(f"C flare {i+1}", fontsize=10)
    axes[i].tick_params(axis='x', rotation=40, labelsize=8)
    axes[i].tick_params(axis='y', labelsize=8)

# hide unused subplot (since 2x6 = 12 but we have 11)
axes[-1].axis("off")

plt.tight_layout()
plt.show()
```

```python
def analyze_flare(t, flux, title):

    # ---------------------------------
    # Convert time → seconds
    # ---------------------------------

    # GOES time is datetime format.
    # Fourier analysis requires evenly spaced numeric time.
    # So we convert time differences into seconds.

    t_seconds = (t - t[0]) / np.timedelta64(1,'s')


    # ---------------------------------
    # 1. LIGHT CURVE
    # ---------------------------------

    plt.figure(figsize=(10,4))
    plt.plot(t, flux)
    plt.title(title + " — Light Curve")
    plt.xlabel("Time")
    plt.ylabel("Flux")
    plt.show()


    # ============================================
    # RAW DATA ANALYSIS (NO DETRENDING)
    # ============================================

    # ---------------------------------
    # 2. FFT on RAW light curve
    # ---------------------------------

    dt = np.mean(np.diff(t_seconds))   # average cadence
    N = len(flux)                      # number of data points

    fft_vals = rfft(flux)              # Fourier transform
    freq = rfftfreq(N, dt)             # frequency axis

    power = np.abs(fft_vals)**2        # power spectrum

    # Remove zero frequency
    
    freq = freq[1:]
    power = power[1:]

    period = 1 / freq
    
    mask = (period>5) & (period<300)   # restrict to useful QPP range

    plt.figure(figsize=(8,4))
    plt.plot(period[mask], power[mask])
    plt.xlabel("Period (seconds)")
    plt.ylabel("Power")
    plt.title(title + " — FFT Spectrum (RAW)")
    plt.show()


    # ---------------------------------
    # 3. Wavelet on RAW light curve
    # ---------------------------------

    scales = np.arange(1,200)

    coeff, freqs = pywt.cwt(flux, scales, 'morl', dt)

    power_w = np.abs(coeff)**2

    plt.figure(figsize=(10,6))

    plt.imshow(power_w,
               extent=[0,len(flux),scales[-1],scales[0]],
               aspect='auto',
               cmap='jet')

    plt.ylabel("Period scale")
    plt.xlabel("Time index")
    plt.title(title + " — Wavelet Power Spectrum (RAW)")
    plt.colorbar(label="Power")

    plt.show()


    # ============================================
    # DETRENDING
    # ============================================

    trend = savgol_filter(flux, 51, 3)

    residual = flux - trend


    plt.figure(figsize=(10,6))

    plt.subplot(211)
    plt.plot(t, flux,label="Original")
    plt.plot(t, trend,label="Trend")
    plt.legend()
    plt.title("Trend Removal")

    plt.subplot(212)
    plt.plot(t, residual)
    plt.axhline(0,ls="--")
    plt.title("Residual")

    plt.tight_layout()
    plt.show()


    # ============================================
    # FFT on RESIDUAL
    # ============================================

    fft_vals = rfft(residual)

    freq = rfftfreq(N, dt)

    power = np.abs(fft_vals)**2

    freq = freq[1:]
    power = power[1:]

    period = 1 / freq

    mask = (period>5) & (period<300)

    plt.figure(figsize=(8,4))
    plt.plot(period[mask],power[mask])
    plt.xlabel("Period (seconds)")
    plt.ylabel("Power")
    plt.title(title + " — FFT Spectrum (Detrended)")
    plt.show()


    # ============================================
    # Wavelet on RESIDUAL
    # ============================================

    coeff, freqs = pywt.cwt(residual, scales, 'morl', dt)

    power_w = np.abs(coeff)**2

    plt.figure(figsize=(10,6))

    plt.imshow(power_w,
               extent=[0,len(residual),scales[-1],scales[0]],
               aspect='auto',
               cmap='jet')

    plt.ylabel("Period scale")
    plt.xlabel("Time index")
    plt.title(title + " — Wavelet Power Spectrum (Detrended)")

    plt.colorbar(label="Power")

    plt.show()


    # ============================================
    # Red Noise Test
    # ============================================

    r1 = np.corrcoef(residual[:-1], residual[1:])[0,1]

    red_noise = (1-r1**2)/(1+r1**2-2*r1*np.cos(2*np.pi*freq))

    plt.figure(figsize=(8,4))

    plt.plot(period[mask], power[mask], label="Observed")
    plt.plot(period[mask], red_noise[mask], "--", label="Red noise")

    plt.legend()

    plt.xlabel("Period (seconds)")
    plt.ylabel("Power")

    plt.title("Red Noise Test")

    plt.show()


    # ============================================
    # EMD
    # ============================================

    emd = EMD()

    imfs = emd(residual)

    n_imf = imfs.shape[0]

    plt.figure(figsize=(10,2*(n_imf+1)))

    plt.subplot(n_imf+1,1,1)
    plt.plot(residual)
    plt.title("Residual")

    for i in range(n_imf):

        plt.subplot(n_imf+1,1,i+2)
        plt.plot(imfs[i])
        plt.title("IMF "+str(i+1))

    plt.tight_layout()
    plt.show()
```

```python
flare_order = [6,7,3,9,4,10]
```

```python
current_flare = 0
```

```python
def run_next_flare():

    global current_flare

    if current_flare >= len(flare_order):
        print("All selected flares analyzed.")
        return

    i = flare_order[current_flare]

    t_flare, flux_flare = flare_windows[i]

    print(f"Analyzing C Flare {i+1}")

    analyze_flare(t_flare, flux_flare, f"C Flare {i+1}")

    current_flare += 1
```

```python
run_next_flare()
```
